### read data & preprocessing

```python
contents_week1 = pd.read_csv('./contents_0802~0808.csv') # 8월 데이터
contents_week2 = pd.read_csv('./contents_0809~0815.csv')
contents_week3 = pd.read_csv('./contents_0816~0822.csv')
contents_week4 = pd.read_csv('./contents_0823~0829.csv')
contents_week5 = pd.read_csv('./contents_0830~0905.csv')
contents_week6 = pd.read_csv('./contents_0906~0912.csv')
contents_week7 = pd.read_csv('./contents_0913~0919.csv')
contents_week8 = pd.read_csv('./contents_0920~0926.csv')
contents_week9 = pd.read_csv('./contents_1004~1010.csv')   # 10월 데이터
contents_week10 = pd.read_csv('./contents_0405~0411.csv')   # 4월 데이터
```

```python
contents_week1 = contents_week1[contents_week1.cnt>=2]
contents_week2 = contents_week2[contents_week2.cnt>=2]
contents_week3 = contents_week3[contents_week3.cnt>=2]
contents_week4 = contents_week4[contents_week4.cnt>=2]
contents_week5 = contents_week5[contents_week5.cnt>=2]
contents_week6 = contents_week6[contents_week6.cnt>=2]
contents_week7 = contents_week7[contents_week7.cnt>=2]
contents_week8 = contents_week8[contents_week8.cnt>=2]
contents_week9 = contents_week9[contents_week9.cnt>=2]
contents_week10 = contents_week10[contents_week10.cnt>=2]
```

```python
contents_week1_pct =  pd.DataFrame(list(np.percentile(contents_week1.cnt, range(0,100))),columns=['data_w1'])
contents_week1_pct.reset_index(level=0, inplace=True)   # index to column
contents_week2_pct =  pd.DataFrame(list(np.percentile(contents_week2.cnt, range(0,100))),columns=['data_w2'])
contents_week2_pct.reset_index(level=0, inplace=True)
contents_week3_pct =  pd.DataFrame(list(np.percentile(contents_week3.cnt, range(0,100))),columns=['data_w3'])
contents_week3_pct.reset_index(level=0, inplace=True)
contents_week4_pct =  pd.DataFrame(list(np.percentile(contents_week4.cnt, range(0,100))),columns=['data_w4'])
contents_week4_pct.reset_index(level=0, inplace=True)
contents_week5_pct =  pd.DataFrame(list(np.percentile(contents_week5.cnt, range(0,100))),columns=['data_w5'])
contents_week5_pct.reset_index(level=0, inplace=True)
contents_week6_pct =  pd.DataFrame(list(np.percentile(contents_week6.cnt, range(0,100))),columns=['data_w6'])
contents_week6_pct.reset_index(level=0, inplace=True)
contents_week7_pct =  pd.DataFrame(list(np.percentile(contents_week7.cnt, range(0,100))),columns=['data_w7'])
contents_week7_pct.reset_index(level=0, inplace=True)
contents_week8_pct =  pd.DataFrame(list(np.percentile(contents_week8.cnt, range(0,100))),columns=['data_w8'])
contents_week8_pct.reset_index(level=0, inplace=True)
contents_week9_pct =  pd.DataFrame(list(np.percentile(contents_week9.cnt, range(0,100))),columns=['data_w9'])
contents_week9_pct.reset_index(level=0, inplace=True)
contents_week10_pct =  pd.DataFrame(list(np.percentile(contents_week10.cnt, range(0,100))),columns=['data_w10'])
contents_week10_pct.reset_index(level=0, inplace=True)
```

```python
df_merge2 =pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(contents_week1_pct, contents_week2_pct, on='index'),contents_week3_pct,on='index'),
                            contents_week4_pct,on='index'),contents_week5_pct,on='index'),contents_week6_pct, on='index'),
                  contents_week7_pct,on='index'),contents_week8_pct,on='index'),contents_week9_pct, on='index'),contents_week10_pct,on='index')
```

## Line-plot

```python
line = (
    Line()
    .add_xaxis(df_merge2.index.tolist())
    .add_yaxis("data_w10",df_merge2.data_w8.tolist(), markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")])) # 4월한주
    .add_yaxis("data_w1",df_merge2.data_w4.tolist(), markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")])) # 8월한주
    .add_yaxis("data_w6",df_merge2.data_w8.tolist(), markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")])) # 9월한주
    .add_yaxis("data_w9",df_merge2.data_w8.tolist(), markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")])) # 10월한주
    .set_global_opts(title_opts=opts.TitleOpts(title="UserCount by percentile"), xaxis_opts=opts.AxisOpts(name="percentile"),yaxis_opts=opts.AxisOpts(name="contents count"),
                    toolbox_opts=opts.ToolboxOpts())
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_colors(["red","blue","green","orange"])
)
line.render()
HTML('render.html')
line.render_notebook()
```

## Pie-chart

```python
# 7월 pie-chart
labels=['noActive','some','Active_low','Active_med','Active_high']

pie = (
    Pie()
    .add("", [list(z) for z in zip(labels,ActiveUser.T.iloc[1:,0].values.tolist())], 
         radius=[30,75], 
         center=["40%", "50%"],
        )
    .set_global_opts(title_opts=opts.TitleOpts(title="Active User_April"), 
                     legend_opts=opts.LegendOpts(orient="vertical", pos_top="15%", pos_left="2%"),toolbox_opts=opts.ToolboxOpts())
    .set_colors(["lightskyblue","lightcoral","yellowgreen","orangered","mediumpurple"])
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
pie.render()
HTML('render.html')
pie.render_notebook()
```
